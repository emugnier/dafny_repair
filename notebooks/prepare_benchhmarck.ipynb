```python
def replace_path(file_path):
    # Extract the file name from the file path
    file_name = file_path.split("/")[-1]

    # Replace the path with './results/'
    new_path = "./results/" + file_name

    return new_path


# Apply the function to 'Original File' and 'New Method File' columns
selected_rows["New Method File"] = selected_rows["New Method File"].apply(replace_path)
print(selected_rows.iloc[0]["Original File"])
print(selected_rows.iloc[0]["New Method File"])
```
